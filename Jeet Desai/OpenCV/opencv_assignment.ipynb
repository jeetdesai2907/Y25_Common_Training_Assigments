## 2. Drawing and Text
**Task:**
1. Create a blank image of size $500 \times 500$ with 3 color channels.
2. Paint a portion of the image red ($[200:300, 300:400]$).

## 4. Image Transformations
**Task:**
1. Create a function `translate(img, x, y)` to shift an image.
2. Create a function `rotate(img, angle, rotPoint=None)` to rotate an image.
3. Apply a flip transformation to `park.jpg`.

```python
# Write your code here
import cv2
from matplotlib import pyplot as plt
image = cv2.imread(r'C:\Users\Jeet Desai\OneDrive\Documents\AUV\opencv_assn\park.jpg')
def translate(image = None, x = 0, y = 0):
    T = np.float32([[1, 0, x], [0, 1,y]])
    h, w = image.shape[:2]
    return cv2.warpAffine(image, T, (w, h))
def rotate(image = None, angle = 0, rotPoint = None):
    if rotPoint is None:
        rotPoint = (image.shape[1] // 2, image.shape[0] // 2)
    rotation_matrix = cv2.getRotationMatrix2D(rotPoint, angle, 1)    
    return cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]))
flipped = cv2.flip(image, 1)
imshow("Horizontal Flip", flipped)
```

## 5. Contour Detection
**Task:**
1. Load `cats.jpg`.
2. Convert to grayscale and apply Gaussian blur.
3. Find the edges using Canny.
4. Use `cv.findContours` to find all contours.
5. Print the count and draw them on a blank image.

```python
# Write your code here
import cv2
from matplotlib import pyplot as plt
import numpy as np
image = cv2.imread(r'C:\Users\Jeet Desai\OneDrive\Documents\AUV\opencv_assn\cat.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
Gaussian = cv2.GaussianBlur(gray_image, (7,7), 0)
imshow('Gaussian Blur', Gaussian)
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
edges = cv2.Canny(Gaussian, 100, 200)
imshow('Canny Edge Detection', edges)
contours, hierarchy = cv2.findContours(
    edges,
    cv2.RETR_LIST,
    cv2.CHAIN_APPROX_SIMPLE
)
print("Number of contours:", len(contours))
blank = np.zeros(image.shape, dtype='uint8')
cv2.drawContours(blank, contours, -1, (0,255,0), 1)
imshow("Contours", blank)
```

## 8. Bitwise Operations
**Task:**
1. Create a $400 \times 400$ blank image.
2. Create a rectangle and a circle on separate blank images.
3. Perform Bitwise AND, OR, XOR, and NOT.

```python
# Write your code here
import cv2
from matplotlib import pyplot as plt
import numpy as np
image = np.zeros((400,400),dtype = np.uint8)
cpy = image.copy()
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
rectangle = cv2.rectangle(image,(100,100),(300,300),(255,255,255),-1)
circle = cv2.circle(cpy,(100,200),90,(255,255,255),-1)
imshow("Rectangle", rectangle)
imshow("Circle", circle)
imshow("AND", cv2.bitwise_and(rectangle, circle))
imshow("OR", cv2.bitwise_or(rectangle, circle))
imshow("XOR", cv2.bitwise_xor(rectangle, circle))
imshow("NOT_rect", cv2.bitwise_not(rectangle))
imshow("NOT_circle", cv2.bitwise_not(circle))
```

## 9. Thresholding
